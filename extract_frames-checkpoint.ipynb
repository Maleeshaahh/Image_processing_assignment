```python
import cv2
import os

# Folder containing videos
video_folder = "Input_video"

# Folder to save frames
frames_folder = "frames"

# Folder to save special frames
keyframes_folder = "keyframes"

# Create folders if they do not exist
os.makedirs(frames_folder, exist_ok=True)
os.makedirs(keyframes_folder, exist_ok=True)

# Loop through each video
for video_file in os.listdir(video_folder):

    video_path = os.path.join(video_folder, video_file)

    print("\nProcessing video:", video_file)

    cap = cv2.VideoCapture(video_path)

    # Check video opened
    if not cap.isOpened():
        print("Video could not open:", video_file)
        continue
    else:
        print("Video opened successfully")

    # Get total number of frames
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    middle_frame = total_frames // 2

    # Calculate step to keep frames around 100
    step = max(1, total_frames // 100)

    print("Total frames:", total_frames)
    print("Frame step:", step)

    frame_count = 0

    while True:

        ret, frame = cap.read()

        if not ret:
            print("Video ended")
            break

        # Convert frame to grayscale
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # Save frames based on calculated step
        if frame_count % step == 0:

            frame_name = os.path.join(
                frames_folder,
                f"{video_file}_frame_{frame_count}.jpg"
            )

            cv2.imwrite(frame_name, gray)
            print("Saved grayscale frame:", frame_name)

        # Save first frame
        if frame_count == 0:

            cv2.imwrite(
                os.path.join(keyframes_folder, f"{video_file}_FIRST.jpg"),
                gray
            )

            print("Saved FIRST frame")

        # Save middle frame
        if frame_count == middle_frame:

            cv2.imwrite(
                os.path.join(keyframes_folder, f"{video_file}_MIDDLE.jpg"),
                gray
            )

            print("Saved MIDDLE frame")

        # Save last frame
        if frame_count == total_frames - 1:

            cv2.imwrite(
                os.path.join(keyframes_folder, f"{video_file}_LAST.jpg"),
                gray
            )

            print("Saved LAST frame")

        frame_count += 1

    cap.release()

    print("Done !:", video_file)

print("\nAll videos processed")
```
