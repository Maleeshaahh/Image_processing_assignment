```python
import cv2
import numpy as np
import os
import random

# Folders
enhanced_folder = "frames"  # Already enhanced frames
output_folder = "enhanced_2"         # Folder to save frames with noise
os.makedirs(output_folder, exist_ok=True)

# Function to add salt and pepper noise
def add_salt_pepper_noise(image, salt_prob=0.01, pepper_prob=0.01):
    noisy = image.copy()
    total_pixels = image.shape[0] * image.shape[1]

    # Salt noise
    num_salt = int(total_pixels * salt_prob)
    for _ in range(num_salt):
        i = random.randint(0, image.shape[0]-1)
        j = random.randint(0, image.shape[1]-1)
        noisy[i, j] = 255

    # Pepper noise
    num_pepper = int(total_pixels * pepper_prob)
    for _ in range(num_pepper):
        i = random.randint(0, image.shape[0]-1)
        j = random.randint(0, image.shape[1]-1)
        noisy[i, j] = 0

    return noisy

# Process all enhanced frames
for filename in os.listdir(enhanced_folder):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        frame_path = os.path.join(enhanced_folder, filename)
        frame = cv2.imread(frame_path, cv2.IMREAD_GRAYSCALE)  # already grayscale
        if frame is None:
            print(f"Cannot read {filename}")
            continue
        noisy_frame = add_salt_pepper_noise(frame, salt_prob=0.02, pepper_prob=0.02)
        save_path = os.path.join(output_folder, f"noisy_{filename}")
        cv2.imwrite(save_path, noisy_frame)
        print(f"Saved noisy frame: {save_path}")

print("All frames with salt-and-pepper noise saved in enhanced_2!")
```
